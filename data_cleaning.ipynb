```python
# Read the first 10,000 rows as a subset
subset_size = 10000
df = pd.read_csv('dataset/enwiki-20170820.csv', nrows=subset_size)
```

```python
# # Replace NaN values in 'SECTION_TEXT' column with empty strings
# df['SECTION_TEXT'] = df['SECTION_TEXT'].fillna('')

# Drop rows with null values in SECTION_TEXT column
column_name = 'SECTION_TEXT'  # Specify the column name with null values you want to consider
df = df.dropna(subset=[column_name])
df.info()
```

```python

# # Group by ARTICLE_ID and TITLE and concatenate SECTION_TEXT
# combined_df = df.groupby(['ARTICLE_ID', 'TITLE'])['SECTION_TEXT'].apply(' '.join).reset_index()

# Group by ARTICLE_ID and concatenate SECTION_TEXT
combined_df = df.groupby('ARTICLE_ID')['SECTION_TEXT'].apply(' '.join).reset_index()

# Rename the aggregated column to ARTICLE_TEXT
combined_df = combined_df.rename(columns={'SECTION_TEXT': 'ARTICLE_TEXT'})
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def clean_and_preprocess_text(text):
    # Step 1: Lowercase the text
    text = text.lower()
    
    # Step 2: Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Step 3: Tokenize the text into words
    words = word_tokenize(text)

    # Remove non-alphabetic characters
    words = [word for word in words if word.isalpha()]
    
    # Step 4: Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    # Remove words of length <= 2
    words = [word for word in words if len(word) > 4]
    
    # Step 5: Lemmatize the words
    # lemmatizer = WordNetLemmatizer()
    # words = [lemmatizer.lemmatize(word) for word in words]
    
    # Join the words back into a single string
    cleaned_text = ' '.join(words)
    
    return cleaned_text

# Example usage:
# text = "This is an example sentence for cleaning and preprocessing. It involves tokenization, removing stopwords, and lemmatization."
text = combined_df.iloc[0, 1]
cleaned_text = clean_and_preprocess_text(text)
# print("Original Text:")
# print(text)
print("\nCleaned and Preprocessed Text:")
print(cleaned_text)
```

```python
combined_df["ARTICLE_TEXT"] = combined_df["ARTICLE_TEXT"].apply(clean_and_preprocess_text)
combined_df["ARTICLE_TEXT"]
```

```python
# Save the DataFrame to a CSV file
# combined_df.to_csv('cleaned_text.csv', index=False)  # Set index=False if you don't want to save row indices

# Save dataframe to text file
with open('cleaned_data.txt', 'w', encoding='utf-8') as f:
    for index, row in combined_df.iterrows():
        f.write(f"{row['ARTICLE_ID']}\t{row['ARTICLE_TEXT']}\n")
```
